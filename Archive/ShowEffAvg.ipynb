```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
from math import floor
```

```python
with open('Archive/avgeffx.pkl', 'rb') as f:
    xs = np.array(pickle.load(f))
with open('Archive/avgeffy.pkl', 'rb') as f:
    ys = np.array(pickle.load(f))
```

```python
vals = np.vstack((xs, ys)).T
vals = vals[vals[:, 0].argsort()]

valsperx = len(xs) / len(set(xs))

nys = [0] * (len(set(xs)))

for i in range(vals.shape[0]):
    nys[floor(i / valsperx)] += vals[i,1]
    
nys = np.array(nys)

nys = nys / (valsperx)

nxs = list(range(min(xs), max(xs)+1, len(set(xs))))
```

```python
plt.scatter(xs, ys, c="blue", s=4)
plt.plot(nxs, nys, c="red")
plt.xlabel("Window size")
plt.ylabel("Std. Dev.")
plt.title("Window vs. Std Dev.")
plt.show()
```
